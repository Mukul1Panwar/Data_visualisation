```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Electric_Vehicle_Population_Data[2].csv")
```

```python
df['County'] = df['County'].fillna(df['County'].mode()[0])
```

```python
df = df.dropna()
```

```python
df.isna().sum().sum()
```

# EV Adoption Over Time: Analyze the growth of the EV population by model year.


```python
import seaborn as sns
```

```python
Y=df['Model Year'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=Y.index,y=Y.values,palette="plasma")
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Adoption of EV over the Year')
plt.show()
```

# Geographical Distribution: Understand where EVs are most commonly registered (e.g., by county or city).


```python
Y=df['City'].value_counts().head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=Y.index,y=Y.values,palette="magma")
```

# EV Types: Breakdown of the dataset by electric vehicle type (BEV, etc.).


```python
Y=df['Electric Vehicle Type'].value_counts()
```

```python
plt.pie(Y.values,labels=Y.index,autopct='%1.2f%%')
plt.show()
```

# Make and Model Popularity: Identify the most popular makes and models among the registered EVs.




```python
df.head()
```

```python
Y=df['Make'].value_counts().head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=Y.values,y=Y.index,palette="viridis")
plt.xlabel("No. of registered EVs")
plt.ylabel("Company Names")
plt.show()
```

# Electric Range Analysis: Analyze the electric range of vehicles to see how EV technology is progressing.


```python
sns.histplot(df['Electric Range'],bins=20,kde=True,color='royalblue')
plt.axvline(df['Electric Range'].mean(),linestyle='--',color='red',label=f'Mean range : {df["Electric Range"].mean():.2f} miles')
plt.legend()
plt.show()
```

# Estimated Growth in Market Size: Analyze and find the estimated growth in the market size of electric vehicles.


```python
df[df['Model Year']==1999]
```

```python
df1 = df.groupby('Model Year')['Electric Range'].mean().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(x='Model Year',y='Electric Range',data=df1,marker='o')
```

# the top 10 models with the highest average electric range


```python
# each model
y = df.groupby('Model')['Electric Range'].mean().sort_values(ascending=False).reset_index().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='Model',y='Electric Range',data=y,palette='cool')
```
